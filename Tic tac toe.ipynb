```python
#----Global Variables-----------

# Game Board
board = ['-', '-', '-',
         '-', '-', '-',
         '-', '-', '-']

# Game is running
game_running = True

# Who wins
winner = None

# Whose turn it is
current_player = 'X'

# Display board
def display_board():
    print(board[0] + '|' + board[1] + '|' + board[2])
    print(board[3] + '|' + board[4] + '|' + board[5])
    print(board[6] + '|' + board[7] + '|' + board[8])

display_board()


# Play a Tic tac toe game    
def play_game():
    
    while game_running:
        
        board_position(current_player)
        if_game_over()
        switch_player()
        
# game is over
    if winner == 'X' or winner == 'O':
        print(winner, 'has won the current game.')
    elif winner == None:
        print('The current game is tied.')
        
        
# Position of a player on board    
def board_position(player):
    print(player,"'s turn")
    position = input('Give a position from 1 to 9:')
    
    # To handle invalid input from a player
    valid = False
    while not valid:
    
        while position not in ['1', '2','3','4','5','6','7','8','9']:
            position = input('Invalid, choose again a position from 1 to 9:')
        
        position = int(position) -1
    
        if board[position] == '-':
            valid = True
        else:
            print('choose an empty position')
    
    board[position] = player    
    
    display_board()
        

# check if game is over
def if_game_over():
    if_winner()
    if_tie()
    

# Who is a winner
def if_winner():
    # set up global variable
    global winner 
    
    row_winner = check_row()
    column_winner = check_col()
    diagonal_winner =check_diagonal()
    
    if row_winner:
         winner = row_winner
    elif column_winner:
         winner = column_winner
    elif diagonal_winner:
         winner = diagonal_winner
    else:
         winner = None
    return


# Check if rows have same values
def check_row():
    global game_running  
    
    row1 = board[0] ==  board[1] == board[2] != '-'
    row2 = board[3] ==  board[4] == board[5] != '-'
    row3 = board[6] ==  board[7] == board[8] != '-'
   
    if row1 or row2 or row3:
        game_running = False
    if row1:
        return board[0]
    elif row2:
        return board[3]
    elif row3:
        return board[6]
    return


# Check if columns have same values
def check_col():
    global game_running  
    
    col1 = board[0] ==  board[3] == board[6] != '-'
    col2 = board[1] ==  board[4] == board[7] != '-'
    col3 = board[2] ==  board[5] == board[8] != '-'
   
    if col1 or col2 or col3:
        game_running = False
    if col1:
        return board[0]
    elif col2:
        return board[1]
    elif col3:
        return board[2]
    return


# Check if diagonals have same values
def check_diagonal():
    # set up global variable
    global game_running  
    
    # check if diagonal have same variables and are not empty
    diagonal1 = board[0] ==  board[4] == board[8] != '-'
    diagonal2 = board[2] ==  board[4] == board[6] != '-'
    
    if diagonal1 or diagonal2:
        game_running = False
    if diagonal1:
        return board[0]
    elif diagonal2:
        return board[2]
    return


# check if game is tied
def if_tie():
    # if there no empty space in board, then it is a tie game
    global game_running
    
    if '-' not in board:
        game_running =False
    return


#Switch player's turn during a game
def switch_player():
    global current_player
    
    # if X then switch to O
    if current_player == 'X':
        current_player = 'O'
    # if O then switch to X
    elif current_player == 'O':
        current_player = 'X'
    return

play_game()
```
